```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import math
```

Parameters from ostrander's gas sand model

```python
Vpseal=10000.0 #in f/s
Vsseal=4082.48 #in f/s
Rhoseal=2.40  #in g/cc
IncAngle=45.0
Vpres=8000.0 #m/s
Vsres=5333.33  #m/s
Rhores=2.14  #in g/cc
deg = np.pi/180 # for rad to deg
```

calculate variables for Aki-Richard & Shuey Aproximation

```python
tou1=((Vpseal/Vsseal)*(Vpseal/Vsseal)-2.0)/(2.0*((Vpseal/Vsseal)*(Vpseal/Vsseal))-2.0)
tou2=((Vpres/Vsres)*(Vpres/Vsres)-2.0)/(2.0*((Vpres/Vsres)*(Vpres/Vsres))-2.0)
tou=(tou1+tou2)/2.0

deltou=tou2-tou1;
delAlpha=(Vpres-Vpseal);
delBeta=(Vsres-Vsseal);
delRho=(Rhores-Rhoseal);
beta=(Vsres+Vsseal)/2.0;
alpha=(Vpres+Vpseal)/2.0;
rho=(Rhores+Rhoseal)/2.0;
TranAngle=math.degrees(math.asin((Vpres/Vpseal)*(math.sin(IncAngle*deg))))
teta=(IncAngle+TranAngle)/2.0;
angle=np.array(np.arange(0,teta))
```

Calculate AVO Response for :

A. Shuey Approximation

```python
Bo=(delAlpha/alpha)/(delAlpha/alpha+delRho/rho);
Ao=Bo-2.0*(1.0+Bo)*(1.0-2.0*tou)/(1.0-tou);
As=0.5*(delAlpha/alpha+delRho/rho);
Bs=As*Ao+deltou/((1.0-tou)*(1.0-tou));
Cs=0.5*delAlpha/alpha;
Rteta1=As+Bs*np.sin(angle*deg)*np.sin(angle*deg)+Cs*((np.tan(angle*deg)*np.tan(angle*deg))-
                                                     (np.sin(angle*deg)*np.sin(angle*deg)));  
```

B. Aki Richard Approximation

```python
Aar=0.5*((delAlpha/alpha)+(delRho/rho))
Bar=(0.5*(delAlpha/alpha)-(4.0*((beta/alpha)**2)*(delBeta/beta))-(2.0*((beta/alpha)**2)*(delRho/rho)))
Car=0.5*delAlpha/alpha;
Rteta2=Aar+Bar*np.sin(angle*deg)*np.sin(angle*deg)+Car*((np.tan(angle*deg)*np.tan(angle*deg))-
                                                        (np.sin(angle*deg)*np.sin(angle*deg)));
```

Plot figure

```python
plt.figure(1,figsize = (11,7),facecolor="white")
shuey, = plt.plot(angle,Rteta1, label="shuey ", linestyle='--', c ='r',marker='x',hold='on')
ar, = plt.plot(angle,Rteta2, label="Aki-Richard ", marker='o', linestyle='--')
plt.ylim(-0.5,0.0)
plt.xlabel('Teta')
plt.ylabel('Rp')
plt.grid()
plt.legend([shuey, ar], ['Shuey ', 'Aki-Richard'], loc=0)
plt.title('Comparison between Aki-Ricard and Shuey \n showing Aki-Ricard amplitude brighter than shuey ', fontsize=16)
plt.show()
```
