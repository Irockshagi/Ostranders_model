Create 3-Layer Model Parameters [Layer1(shale), Layer2(Gas Sand), Layer 3(Shale)] in pyhton array as pict below


```python
vp_mod = [10000.0, 8000.0, 10000.0]  # P-wave velocity (f/s)
rho_mod = [2.4, 2.14, 2.4]         # Density (g/cc)
pr_mod = [0.4, 0.1, 0.4]   # Poisson Ratio
```

##### calculate Vs model from poisson ratio equation :  $  \sigma =\frac { (\frac{Vp}{Vs} )^2-2}{2[(\frac{Vp}{Vs})^2 -1]}$

```python
vs_mod = [] # zero array for vs_mode
vsa = [] # declare Dummy array for vsa
for i in range(0,3):   
    vsa  = vp_mod[i] / np.sqrt((2.0*pr_mod[i]-2.0)/(2.0*pr_mod[i]-1.0))
    vs_mod.append(vsa )
```
